# **Resize image to 200px 180px ( Width HeightÂ )**

```python
!curl -o Robo.jpg blob:https://colab.research.google.com/47b259a1-5215-4c86-936f-750265b2
import cv2
img = cv2.imread('Robo.jpg',1)
resize = cv2.resize(img, (200, 180))
cv2_imshow(resize)
```

# **Rotating theÂ Image**

```python
!curl -o Robo.jpg blob:https://colab.research.google.com/47b259a1-5215-4c86-936f-750265b2
import cv2
img = cv2.imread('Robo.jpg',1)
h,w = img.shape[:2]
center = (w // 2, h // 2) 
matrix = cv2.getRotationMatrix2D(center, 45, 0.5)  
rotated = cv2.warpAffine(img, matrix, (w, h)) 
cv2_imshow(rotated)
```

# **Drawing a Rectangle onÂ image**

```python
!curl -o Robo.jpg blob:https://colab.research.google.com/47b259a1-5215-4c86-936f-750265b2
import cv2
img = cv2.imread('Robo.jpg',1)
output = img.copy() 
rectangle = cv2.rectangle(output, (400, 400),(800, 200), (0, 255, 0), 5) 

cv2_imshow(rectangle)
```

# **Displaying text on Image in Python usingÂ OpenCV**

```python
!curl -o Robo.jpg blob:https://colab.research.google.com/47b259a1-5215-4c86-936f-750265b2
import cv2
img = cv2.imread('Robo.jpg',1)
output = img.copy() 
text = cv2.putText(output, 'www.Robo.org.in', (120, 25),  cv2.FONT_HERSHEY_SIMPLEX, 1, (60, 60, 60), 2)

cv2_imshow(text)
```
